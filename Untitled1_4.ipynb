```python
from scipy.interpolate import RBFInterpolator
import numpy as np
import matplotlib.pyplot as plt

def estimate_density(samples, centers, bins=100):
    hist, edges = np.histogram(samples, bins=bins, density=True)
    centers_hist = 0.5 * (edges[1:] + edges[:-1])
    density = np.interp(centers, centers_hist, hist)
    return density


normal_samples = np.random.normal(0, 1, size=800000)
uniform_samples = np.random.uniform(-3, 3, size=800000)
num_centers=10000
centers = np.linspace(-10, 10, num_centers)

normal_density = estimate_density(normal_samples, centers)
uniform_density = estimate_density(uniform_samples, centers)


rbf_normal = RBFInterpolator(centers.reshape(-1, 1), normal_density, kernel="multiquadric", epsilon=5)
rbf_uniform = RBFInterpolator(centers.reshape(-1, 1), uniform_density, kernel="multiquadric", epsilon=5)


def interpolate_densities(t, centers):

    normal_weights = rbf_normal(centers.reshape(-1, 1))
    uniform_weights = rbf_uniform(centers.reshape(-1, 1))
    return (1 - t) * normal_weights + t * uniform_weights

t_values = np.linspace(0, 1, 5)
x_plot = np.linspace(-5, 5, 1000).reshape(-1, 1)

plt.figure(figsize=(15, 3))
for i, t in enumerate(t_values):
    y_plot = interpolate_rbf(t, x_plot)

    plt.subplot(1, len(t_values), i+1)
    plt.plot(x_plot, y_plot)
    plt.title(f'Interpolation at t = {t:.2f}')
    plt.ylim(0, 1.2)
    plt.xlim(-5, 5)

plt.tight_layout()
plt.show()

```
